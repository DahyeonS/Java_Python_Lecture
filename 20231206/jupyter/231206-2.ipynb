```python
num = lis[0].find(class_='link_story').attrs['href'].split('movieId=')[1]
title = lis[0].find(class_='link_txt').text
grade = lis[0].find(class_='txt_grade').text
rate = lis[0].find_all(class_='txt_num')[0].text
date = lis[0].find_all(class_='txt_num')[1].text

print(f'번호: {num}, 영화이름: {title}, 평점: {grade}, 예매율: {rate}, 개봉일자: {date}')
print({num:(title,grade,rate,date)})
```

```python
nums = []
titles = []
grades = []
rates = []
dates = []
infos = []

for li in lis :
    num = li.find(class_='link_story').attrs['href'].split('movieId=')[1]
    title = li.find(class_='link_txt').text
    grade = li.find(class_='txt_grade').text
    rate = li.find_all(class_='txt_num')[0].text
    date = li.find_all(class_='txt_num')[1].text
    
    nums.append(num)
    titles.append(title)
    grades.append(grade)
    rates.append(rate)
    dates.append(date)

    infos.append({num:(title,grade,rate,date)})
```

```python
for i in range(len(titles)) :
    print(f'번호: {nums[i]}, 영화이름: {titles[i]}, 평점: {grades[i]}, 예매율: {rates[i]}, 개봉일자: {dates[i]}')
```

```python
for info in infos :
    print(info)
```

### 알라딘 도서검색

```python
url = 'https://www.aladin.co.kr/search/wsearchresult.aspx'
params = {
    'SearchTarget':'All', 'SearchWord':'에이트'
}

response = requests.get(url, params=params)
```

```python
raw_data = response.text
soup = BeautifulSoup(raw_data)
```

```python
booklist = soup.find_all(class_='ss_book_box')[0]

id = booklist.attrs['itemid']
price = booklist.find(class_='ss_book_list').find_all('li')[2].find('span').text
discount = booklist.find(class_='ss_p2').find('b').find('span').text

book_info = booklist.find_all('li')[1].text.split(' | ')

author = book_info[0].split(' (')[0]
publisher = book_info[1]
date = book_info[2]

{id:(author, price, discount, publisher, date)}
```

```python
booklist = soup.find_all(class_='ss_book_box')[:25]
book_infos = []
book_info

for book in booklist :
    id = book.attrs['itemid']
    discount = book.find(class_='ss_p2').find('b').find('span').text
    
    if book.find(class_='book_label_wrap') != None :
        if book.find(class_='ss_ht1') != None :
            book_info = book.find_all(class_='ss_book_list')[0].find_all('li')[3].text.split(' | ')
        else : 
            book_info = book.find_all(class_='ss_book_list')[0].find_all('li')[2].text.split(' | ')
        price, discount = discount, 'None'
    else :
        book_info = book.find_all(class_='ss_book_list')[0].find_all('li')[1].text.split(' | ')
        price = book.find(class_='ss_book_list').find_all('li')[2].find('span').text
    
    author = book_info[0].split(' (')[0]
    publisher = book_info[1]
    date = book_info[2]
    
    book_infos.append({id:(author, price, discount, publisher, date)})

for book_info in book_infos :
    print(book_info)
```
