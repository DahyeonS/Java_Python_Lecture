```python
import requests

url = 'http://apis.data.go.kr/B552584/UlfptcaAlarmInqireSvc/getUlfptcaAlarmInfo'
params = {
    'serviceKey' : skey,
    'returnType' : 'json',
    'numOfRows' : '100',
    'pageNo' : '1',
    'year' : '2023',
    'itemCode' : 'PM10'
}

response = requests.get(url, params=params)
response.status_code
```

```python
raw_json = response.json()
```

```python
item = raw_json['response']['body']['items']

districtname = [i['districtName'] for i in item]
datadate = [i['dataDate'] for i in item]
issuegbn = [i['issueGbn'] for i in item]

districtname, datadate, issuegbn
```
