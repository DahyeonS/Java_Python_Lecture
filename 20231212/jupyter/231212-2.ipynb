```python
# 크롤링 전체 코드
name_list = []
tel_list = []
add_list = []

url = 'https://www.wikitree.co.kr/articles/217101'
response = requests.get(url)
soup = BeautifulSoup(response.text)

for s in soup.select('strong') :
    if '회' in s.text :
        name_list.append(s.text.split('회 ')[-1])

p = soup.select('p')

for i in range(len(p)) :
    if match('[0-9]{2,}-[0-9]{3,}-[0-9]{4}', p[i].text) :
        tel_list.append(p[i].text.split(' ')[0])
        add_list.append(p[i+1].text)

for i in range(len(add_list)) :
    if add_list[i][-1] == ' ' :
        add_list[i] = add_list[i][:-1]

rinfo = list(zip(add_list, tel_list))
rinfos = list(zip(name_list, rinfo))
rest_infos = dict(rinfos)
rest_infos
```

```python
# 좌표
x_list = []
y_list = []

for add in add_list :
    url = 'https://dapi.kakao.com/v2/local/search/address.json'
    params = {
        'query':add
    }
    headers = {
        "Authorization":"KakaoAK REST_API_KEY"
    }
    response = requests.get(url, params=params, headers=headers)
    json_data = response.json()
    
    address = json_data['documents'][0]['address']
    x_list.append(address['x'])
    y_list.append(address['y'])

    time.sleep(0.5)

ainfo = list(zip(x_list, y_list))
ainfos = list(zip(name_list, ainfo))
add_infos = dict(ainfos)
add_infos
```

```python
map = folium.Map(('37.5511225714939', '126.987867837993'), zoom_start=11) # 남산타워 기준

for i in range(len(y_list)) :
    location = (y_list[i], x_list[i])
    text = f'{name_list[i]}<br>{add_list[i]}<br>{tel_list[i]}<br>'
    popup = folium.Popup(text, min_width=50, max_width=200)
    
    folium.Marker(
        location, 
        popup, 
        icon=folium.Icon(color='pink', icon='glyphicon-road')
    ).add_to(map)
```
