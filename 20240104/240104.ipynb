```python
data_dict = {'data_x':[1, 2, 3, 4, 5], 'data_y':[2, 3, 5, 10, 8]}
plt.plot('data_x', 'data_y', data=data_dict)
plt.xlabel('data_x', labelpad=15, fontdict={'family':'serif', 'color':'b', 'weight':'bold', 'size':20})
plt.ylabel('data_y', labelpad=15)
plt.show()
```

```python
data_dict = {'data_x':[1, 2, 3, 4, 5], 'data_y':[2, 3, 5, 10, 8]}
font_y = {
    'family':'serif', 'color':'r', 'weight':'bold', 'size':20
}
plt.plot('data_x', 'data_y', data=data_dict)
plt.xlabel('data_x', labelpad=15, fontdict={'family':'serif', 'color':'b', 'weight':'bold', 'size':20})
plt.ylabel('data_y', labelpad=15, fontdict=font_y)
plt.show()
```

```python
data_dict = {'data_x':[1, 2, 3, 4, 5], 'data_y':[2, 3, 5, 10, 8]}
font_x= {
    'family':'serif', 'color':'b', 'weight':'bold', 'size':20
}
font_y = {
    'family':'serif', 'color':'r', 'weight':'bold', 'size':20
}
plt.plot('data_x', 'data_y', data=data_dict)
plt.xlabel('data_x', labelpad=15, fontdict=font_x, loc='right')
plt.ylabel('data_y', labelpad=15, fontdict=font_y,  loc='top')
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 4, 4], '-', color='C0', label='Solid')
plt.plot([1, 2, 3], [3, 3, 3], '--', color='C0', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], ':', color='C0', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-.', color='C0', label='Dash-Dot')
plt.legend(loc='upper center', ncol=4)
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', color='C0', label='Solid')
plt.plot([1, 2, 3], [3, 3, 3], linestyle='dashed', color='C0', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', color='C0', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', color='C0', label='Dash-Dot')
plt.legend(loc='upper center', ncol=4)
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid',
         linewidth=10, solid_capstyle='butt',
         color='C0', label='Solid')
plt.plot([1, 2, 3], [3, 3, 3], linestyle='dashed', 
         linewidth=10, solid_capstyle='round',
         color='C0', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', 
         linewidth=10, dash_capstyle='round',
         color='C1', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', 
         linewidth=10, dash_capstyle='round',
         color='C1', label='Dash-Dot')
plt.legend(loc='upper center', ncol=2)
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.show()
```
